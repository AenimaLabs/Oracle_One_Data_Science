```python
import pandas as pd

ruta = "/content/drive/MyDrive/Data Science Oracle One/canadian_immegration_data.csv"
inmigracion = pd.read_csv(ruta)

display(inmigracion)
```

```python
inmigracion.isnull().sum() #isna().sum() hace lo mismo, is not available
```

```python
inmigracion.shape
```

```python
inmigracion.info()
```

```python
inmigracion.set_index("Country", inplace=True) #el indice es ahora el pa√≠s
```

```python
display(inmigracion)
```

```python
years = list(map(str, range(1980, 2014))) #incluye el primer n√∫mero, excluye el segundo n√∫mero
display(years)
```

```python
colombia = inmigracion.loc["Colombia", years]
display(colombia)
```

```python
# col_dict = colombia.to_dict()
col_dict = {"A√±o": colombia.index.to_list(),
            "Inmigrantes": colombia.values.tolist()}
display(col_dict)
```

```python
datos_col = pd.DataFrame(col_dict)
display(datos_col)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5)) #tama√±o de la figura 10 pulgadas por 5 pulgadas
plt.plot(datos_col["A√±o"], datos_col["Inmigrantes"]) #plot hace un gr√°fico de l√≠nea
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2013'])
plt.title("Inmigrantes de Colombia a Canad√°")
plt.xlabel("A√±o")
plt.ylabel("Inmigrantes")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5)) #fig es la figura (tama√±o del lienzo) y ax es la gr√°fica
ax.plot(datos_col["A√±o"], datos_col["Inmigrantes"])
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #los ticks van de 5 en 5 de manera autom√°tica
ax.set_title("Inmigrantes de Colombia a Canad√°\n1980 - 2013")
ax.set_xlabel("A√±o")
ax.set_ylabel("Inmigrantes")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,3))
axs[0].plot(datos_col["A√±o"], datos_col["Inmigrantes"])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title("Inmigrantes de Colombia a Canad√°\n1980 - 2013")
axs[0].set_xlabel("A√±o")
axs[0].set_ylabel("Inmigrantes")
axs[0].grid(True)

axs[1].boxplot(datos_col["Inmigrantes"])
axs[1].set_title("Distribuci√≥n de los datos")
axs[1].set_ylabel("Inmigrantes")
axs[1].set_xlabel("Colombia")
axs[1].grid(True)

plt.show()


plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle("Distribuci√≥n de los datos", fontsize=16)

axs[0, 0].plot(inmigracion.loc["Colombia", years])
axs[0, 0].set_title("Inmigrantes de Colombia a Canad√°\n1980 - 2013")


axs[0, 1].plot(inmigracion.loc["Brazil", years])
axs[0, 1].set_title("Inmigrantes de Brasil a Canad√°\n1980 - 2013")


axs[1, 0].plot(inmigracion.loc["Argentina", years])
axs[1, 0].set_title("Inmigrantes de Argentina a Canad√°\n1980 - 2013")


axs[1, 1].plot(inmigracion.loc["Chile", years])
axs[1, 1].set_title("Inmigrantes de Chile a Canad√°\n1980 - 2013")

ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)




for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel("A√±o")
    ax.set_ylabel("Inmigrantes")
    ax.grid(True)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 3)) #fig es la figura (tama√±o del lienzo) y ax es la gr√°fica
ax.plot(datos_col["A√±o"], datos_col["Inmigrantes"], lw=2, marker="o")
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #los ticks van de 5 en 5 de manera autom√°tica
ax.set_title("Inmigrantes de Colombia a Canad√°\n1980 - 2013", fontsize=20, loc="left")
ax.set_xlabel("A√±o", fontsize=14)
ax.set_ylabel("Inmigrantes",fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.grid(True)
plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
ventas_tiendas = pd.DataFrame(ventas_2022, index=tiendas)
ventas_tiendas
```

```python
# Crear la figura y los subgr√°ficos

fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar los espaciados entre los subgr√°ficos

plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Agregar un t√≠tulo general para los subgr√°ficos

fig.suptitle('Ventas en el per√≠odo de enero a diciembre de 2022 en las tiendas A, B, C y D')

# Agregar los gr√°ficos en cada uno de los subgr√°ficos

axs[0, 0].plot(ventas_tiendas.loc['A'])
axs[0, 0].set_title('Ventas en la tienda A')
axs[0, 1].plot(ventas_tiendas.loc['B'])
axs[0, 1].set_title('Ventas en la tienda B')
axs[1, 0].plot(ventas_tiendas.loc['C'])
axs[1, 0].set_title('Ventas en la tienda C')
axs[1, 1].plot(ventas_tiendas.loc['D'])
axs[1, 1].set_title('Ventas en la tienda D')

# Agregar etiquetas para los ejes X y Y

for ax in axs.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('N√∫mero de ventas')

# Mostrar la figura

plt.show()
```
