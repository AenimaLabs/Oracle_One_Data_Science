```python
def generar_grafico (palette):
  fig, ax = plt.subplots(figsize=(10, 3)) #fig es la figura (tama√±o del lienzo) y ax es la gr√°fica
  ax = sns.barplot(data=top_10, x='Total', hue=top_10.index, orient="h", palette=palette)
  ax.set_title('Top 10 pa√≠ses con m√°s inmigraci√≥n hacia Canad√°\n1980 a 2013', fontsize=20, loc='left')
  ax.set_xlabel('N√∫mero de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
def generar_grafico (palette):
  fig, ax = plt.subplots(figsize=(10, 3)) #fig es la figura (tama√±o del lienzo) y ax es la gr√°fica
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis = "y", which = "both", size = 0)
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index , hue=top_10.index, orient="h", palette=palette, legend=False)
  ax.set_title('Top 10 pa√≠ses con m√°s inmigraci√≥n hacia Canad√°\n1980 a 2013', fontsize=20, loc='left')
  ax.set_xlabel('N√∫mero de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  for i, j in enumerate(top_10['Total']):
    ax.text(j, i, str(j), color = "black",ha='left', va='center', fontsize=12)
  plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
ventas_tiendas = pd.DataFrame(ventas_2022, index=tiendas)
ventas_tiendas
```

```python
# Crear la figura y los subgr√°ficos

fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar los espaciados entre los subgr√°ficos

plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Agregar un t√≠tulo general para los subgr√°ficos

fig.suptitle('Ventas en el per√≠odo de enero a diciembre de 2022 en las tiendas A, B, C y D')

# Agregar los gr√°ficos en cada uno de los subgr√°ficos

axs[0, 0].plot(ventas_tiendas.loc['A'])
axs[0, 0].set_title('Ventas en la tienda A')
axs[0, 1].plot(ventas_tiendas.loc['B'])
axs[0, 1].set_title('Ventas en la tienda B')
axs[1, 0].plot(ventas_tiendas.loc['C'])
axs[1, 0].set_title('Ventas en la tienda C')
axs[1, 1].plot(ventas_tiendas.loc['D'])
axs[1, 1].set_title('Ventas en la tienda D')

# Agregar etiquetas para los ejes X y Y

for ax in axs.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('N√∫mero de ventas')

# Mostrar la figura

plt.show()
```

```python
# Crear la figura y los subgr√°ficos
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar los espaciados entre los subgr√°ficos
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Agregar un t√≠tulo general a los subgr√°ficos
fig.suptitle('Ventas en el periodo de enero a diciembre de 2022 en las tiendas A, B, C y D', fontsize=20)

# Lista de colores
colores = ['darkviolet', 'green', 'darkblue', 'coral']

# Bucle para trazar y personalizar los subgr√°ficos
for i, ax in enumerate(axs.flat):
    ax.plot(ventas_tiendas.loc[ventas_tiendas.index[i]], color=colores[i], lw=3)
    ax.set_title(f'Ventas en la tienda {ventas_tiendas.index[i]}', loc='left', fontsize=16)
    ax.set_xlabel('Mes', fontsize=14)
    ax.set_ylabel('N√∫mero de ventas', fontsize=14)
    ax.tick_params(labelsize=12)
    ax.grid(color='lightgrey')
```
