```python
import pandas as pd

ruta = "/content/drive/MyDrive/Data Science Oracle One/canadian_immegration_data.csv"
inmigracion = pd.read_csv(ruta)

display(inmigracion)
```

```python
inmigracion.isnull().sum() #isna().sum() hace lo mismo, is not available
```

```python
inmigracion.shape
```

```python
inmigracion.info()
```

```python
inmigracion.set_index("Country", inplace=True) #el indice es ahora el país
```

```python
display(inmigracion)
```

```python
years = list(map(str, range(1980, 2014))) #incluye el primer número, excluye el segundo número
display(years)
```

```python
colombia = inmigracion.loc["Colombia", years]
display(colombia)
```

```python
# col_dict = colombia.to_dict()
col_dict = {"Año": colombia.index.to_list(),
            "Inmigrantes": colombia.values.tolist()}
display(col_dict)
```

```python
datos_col = pd.DataFrame(col_dict)
display(datos_col)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5)) #tamaño de la figura 10 pulgadas por 5 pulgadas
plt.plot(datos_col["Año"], datos_col["Inmigrantes"]) #plot hace un gráfico de línea
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2013'])
plt.title("Inmigrantes de Colombia a Canadá")
plt.xlabel("Año")
plt.ylabel("Inmigrantes")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5)) #fig es la figura (tamaño del lienzo) y ax es la gráfica
ax.plot(datos_col["Año"], datos_col["Inmigrantes"])
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #los ticks van de 5 en 5 de manera automática
ax.set_title("Inmigrantes de Colombia a Canadá")
ax.set_xlabel("Año")
ax.set_ylabel("Inmigrantes")
plt.show()
```

# Práctica

```python
df_comparacion = inmigracion.loc[['Brazil', 'Argentina'], years]
df_comparacion
```

```python
df_comparacion = df_comparacion.T #T de transpose
df_comparacion
```

```python
plt.plot(df_comparacion['Brazil'], label='Brazil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```
